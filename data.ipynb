```python
# Remove end spaces
def remove_end_spaces(string):
    return "".join(string.rstrip())
```

```python
def convert_data(file, pdf2):
    df= pd.read_csv(path+str(file))
    name=str(file).split('.',1)[0]
    #print(name)
    d={'image_id':str(name), 'labels':(df["unicode"]+ ' '+ df["a"].astype(str)+ ' '+df["b"].astype(str)+
                                       ' '+df["c"].astype(str)+ ' '+df["d"].astype(str)+ ' ')}
    pdf= pd.DataFrame(data=d)
    a=pdf['labels'].tolist()
    a=remove_end_spaces(''.join(map(str, a)))
    #print(a)
    pdf= pd.DataFrame({"image_id":name,'labels':[a]})
    pdf2= pdf2.append(pdf, ignore_index=True)
    d.clear()  
    del a
    return pdf2
```

```python
path='Add_unicode_label_csv/'
dataa= pd.DataFrame({"image_id":[],'labels':[]})
files = (file for file in os.listdir(path) 
         if os.path.isfile(os.path.join(path, file)))
for file in files:
    dataa= convert_data(file, dataa)
dataa
#Solved!!!
#ToDo: Se agrega un espacio en blanco al final de la lista de "labels" que debe ser removido malualmente!!!
#Se corrige este detalle al final de este notebook una vez creado data/train, test dataset, tambien se corrige en el merge_csv.csv
```
